```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 定义五个维度标签和数据
labels = ['extreme high T', 'extreme hot days', 'continuous heat', 'heat days', 'continuous hot days']
data_2003 = [2, 3, 4, 3, 2]
data_2017 = [3, 4, 5, 4, 3]
data_2006 = [1, 2, 3, 2, 1]
data_2022 = [4, 5, 4, 5, 4]
data_2024 = [5, 5, 5, 5, 5]

# 2. 计算每个顶点的角度（五边形均匀分布）
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

# 3. 添加第一个点到末尾，使得图形闭合
data_2003 += data_2003[:1]
data_2017 += data_2017[:1]
data_2006 += data_2006[:1]
data_2022 += data_2022[:1]
data_2024 += data_2024[:1]
angles += angles[:1]

# 4. 创建图形
fig, ax = plt.subplots(figsize=(6, 6), dpi=100, subplot_kw=dict(polar=True))

# 5. 绘制每个年份的数据
ax.plot(angles, data_2003, color='cyan', linewidth=2, label='2003')
ax.plot(angles, data_2017, color='orange', linewidth=2, label='2017')
ax.plot(angles, data_2006, color='green', linewidth=2, label='2006')
ax.plot(angles, data_2022, color='red', linewidth=2, label='2022')
ax.plot(angles, data_2024, color='blue', linewidth=2, label='2024')
 
# 6. 填充区域
ax.fill(angles, data_2003, color='cyan', alpha=0.25)
ax.fill(angles, data_2017, color='orange', alpha=0.25)
ax.fill(angles, data_2006, color='green', alpha=0.25)
ax.fill(angles, data_2022, color='red', alpha=0.25)
ax.fill(angles, data_2024, color='blue', alpha=0.25)

ax.set_yticklabels([])  # 隐藏半径标签
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12, fontweight='bold')

ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1))



plt.tight_layout()
plt.show()
```
